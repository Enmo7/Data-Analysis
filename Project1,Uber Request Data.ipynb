### What is the busiest day for orders in General? 

#### Change data type for (Drop timestamp, Request timestamp)

```python
# The Original data
df
```

```python
# Change data type for (Drop timestamp, Request timestamp)
df['Request timestamp'] = pd.to_datetime(df['Request timestamp'])
df['Drop timestamp'] = pd.to_datetime(df['Drop timestamp'])
```

#### Separating date from time:

The date will be separated from the time, each in a special column, in order to find the busiest day.

```python
df['new_date_Request_timestamp'] = [d.date() for d in df['Request timestamp']]
df['new_time_Request_timestamp'] = [d.time() for d in df['Request timestamp']]
df
```

#### Change data type for ( new_date_Request_timestamp) column:

```python
df['new_date_Request_timestamp'] = pd.to_datetime(df['new_date_Request_timestamp'])
```
